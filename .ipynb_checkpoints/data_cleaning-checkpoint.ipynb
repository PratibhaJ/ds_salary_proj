### Salary Parsing

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x : 1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

```python
# remove text part of salary estimate field
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary.head()
```

```python
# remove 'k','$','-' characters from salary field
# minus k and dollar symbol
minus_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
minus_kd.head()
```

```python
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
# 53-91 , min and max salary save differently
df['min_salary'] = min_hr.apply(lambda x : x.split('-')[0])
```

```python
df.head(2)
```

```python
df['max_salary'] = min_hr.apply(lambda x : x.split('-')[1])
```

```python
df.head(2)
```

```python
# with min and max salary values we can calculate average salary
cols = ['min_salary','max_salary']
df['avg_salary'] = df[cols].astype(float).mean(axis=1)
```

```python
df.head(2)
```

### Company name text only

```python
df['company_txt'] = df.apply(lambda x : x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3], axis=1)
```

```python
df.head(2)
```

### State Field

```python
# df['job_state'] = df['Location'].apply(lambda x: x.split(',')[0])
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.job_state.value_counts()
```

```python
df.job_state.value_counts()
```

```python
# If location and headquarters are same make it as 1, else 0
df['same_state'] = df.apply(lambda x: 1 if x.Location==x.Headquarters else 0, axis=1)
```

```python
# Age of the company
df['age'] = df.Founded.apply(lambda x: x if x<1 else 2020-x)
```

```python
df.head(2)
```

```python
# Parsing of Job description with data science keywords like: python, spark, r studio, aws, excel

#python
df['python_yn'] = df['Job Description'].apply(lambda x : 1 if 'python' in x.lower() else 0)
df.python_yn.value_counts()
```

```python
# R studio
df['R_yn'] = df['Job Description'].apply(lambda x : 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()
```

```python
# Spark
df['spark'] = df['Job Description'].apply(lambda x : 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()
```

```python
# AWS
df['aws'] = df['Job Description'].apply(lambda x : 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()
```

```python
# Excel
df['excel'] = df['Job Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
```

```python
# remove first two columns
df.columns
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.drop(['Unnamed: 0.1'],axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.to_csv('salary_data_cleaned.csv',index=False)
```
